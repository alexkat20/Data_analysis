4) Убираем строки с пустыми ячейками. 
Здесь находим такие строки, где все 4 колонки Age, Height, Weight, Medal пустые, так как кажется,  что в них нет смысла для дальнейшей работы. Это одни из самых важных ячеек для работы с данными. Без них дальнейший анализ будет не до конца правильным.

```python
#  data[data['Age'].isna()][data['Height'].isna()][data['Weight'].isna()]
data1 = data.drop(data[data['Age'].isna()][data['Height'].isna()][data['Weight'].isna()][data['Medal'].isna()].index)
data1.isna().sum()
#  data1[data1['Age'].isna()][data1['Height'].isna()][data1['Weight'].isna()]
```

5) Заполняем пустые ячейки возраста медианой всех значений возраста. Использование среднего значения будет некорректным, так как на нужен целый возраст, а при выявлении среднего значения возраст может быть дробным что не подходит по условию.

```python
#  data['Age'] = data['Age'].fillna(data['Age'].mean())
#  data.isna().sum()

data1['Age'] = data1['Age'].fillna(data1['Age'].median())
data1.isna().sum()
```

6) Заменяем пустые ячейки роста на средний рост. Тут можно использовать дробные числа, поэтому подходит среднее значение по росту всех людей в таблице

```python
#  data['Height'] = data['Height'].fillna(data['Height'].mean())
#  data.isna().sum()

data1['Height'] = data1['Height'].fillna(data1['Height'].mean())

data1.isna().sum()
```

7) По аналогии с ростом заполняем вес средним значением, так как он тоже может быть дробным

```python
#  data['Weight'] = data['Weight'].fillna(data['Weight'].mean())
#  data.isna().sum()

data1['Weight'] = data1['Weight'].fillna(data1['Weight'].mean())
data1.isna().sum()
```

8) С медалями сложнее. Ни мода, ни медиана, ни среднее не подходят, так как тут только строки представлены, поэтому заменим пустоту на слово Nothing, что означает "ничего"

```python
#  data['Medal'] = data['Medal'].fillna('Nothing')
#  data.isna().sum()

data1['Medal'] = data1['Medal'].fillna('Nothing')
data1.isna().sum()
data1.sample(10)
```
