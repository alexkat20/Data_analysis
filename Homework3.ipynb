<h2> Так как первый датасет совсем небольшой, а во втором чуть больше 800 строчек, то объядиняем два датасета в один для большего количества данных, на которых будем обучаться</h2>

```python
new_data = pd.concat([data,data1], axis=0, ignore_index=True)
```

```python
new_data
```

<h2>Самое простое - совместить две колоник с ценами, так как здесь только числа. Заполняем все пустые значения одной колонки значениями другой</h2>

```python
new_data['Price'] = new_data['Price'].fillna(new_data['price'])
```

```python
new_data
```

<h2>Возьмем за основу вторую таблицу, в которой строк больше. Там всего 12 колонок, на них и будем ориентироваться. Удалим все ненужные колонки</h2>

```python
new_data = new_data.drop(columns=['car_ID',	'symboling', 'fueltype',	'aspiration', 'enginelocation',	'wheelbase',	'carlength',	'carwidth',	'carheight',	'curbweight',	'enginetype', 'enginesize',	'fuelsystem',	'boreratio',	'stroke',	'compressionratio',	'horsepower',	'peakrpm',	'citympg',	'highwaympg',	'price'])
```

```python
new_data
```

<h2>Заполняем столбец с названием производителей. Для этого нужно разделить столбец carname на части и взять нулевую, так как там название фирмы производителя</h2>

```python
new_data['Make'] = new_data['Make'].fillna(new_data['CarName'].str.split(' ', expand=True)[0])
```

<h2>Создаем временный датафрейм, из которого соберем модель каждой машины</h2>

```python
tmp_data = new_data['CarName'].str.split(' ', expand=True)
tmp_data = tmp_data.fillna(' ')
tmp_data

```

<h2>Заполняем пустые ячейки моделей машин новыми значениями из временного датафрейма</h2>

```python
new_data['Model'] = new_data['Model'].fillna(tmp_data[1] + ' ' + tmp_data[2] + ' ' + tmp_data[3] + ' ' + tmp_data[4])
```

```python
new_data
```

<h2>Заполняем пропуски в столбце количества дверей. Здесь разные типы данных - строки и числа с плавающей точкой. Оказалось, что в столбце всего два варианта строк - two, four. Заменим их на числа, чтобы столбец был одного типа</h2>

```python
new_data['Doors'] = new_data['Doors'].fillna(new_data['doornumber'])

new_data.loc[new_data['Doors'] == 'two', 'Doors'] = 2
new_data.loc[new_data['Doors'] == 'four', 'Doors'] = 4
new_data = new_data.drop(columns=['CarName', 'doornumber'])
new_data
```

<h2>По аналогии поступаем со столбцом Cylinder. Здесь больше значений, но это не очень страшно</h2>

```python
print(new_data['cylindernumber'].value_counts())

new_data['Cylinder'] = new_data['Cylinder'].fillna(new_data['cylindernumber'])

new_data.loc[new_data['Cylinder'] == 'two', 'Cylinder'] = 2
new_data.loc[new_data['Cylinder'] == 'four', 'Cylinder'] = 4
new_data.loc[new_data['Cylinder'] == 'six', 'Cylinder'] = 6
new_data.loc[new_data['Cylinder'] == 'five', 'Cylinder'] = 5
new_data.loc[new_data['Cylinder'] == 'eight', 'Cylinder'] = 8
new_data.loc[new_data['Cylinder'] == 'twelve', 'Cylinder'] = 12
new_data.loc[new_data['Cylinder'] == 'three', 'Cylinder'] = 3
new_data = new_data.drop(columns=['cylindernumber'])
new_data
```

<h2>Заполняем пропуски в типе машин на значения carbody из первой таблицы</h2>

```python
new_data['Type'] = new_data['Type'].fillna(new_data['carbody'])
new_data = new_data.drop(columns=['carbody'])
new_data
```

<h2>Удаляем столбцы, которые сложно соотнести </h2>

```python
new_data = new_data.drop(columns=['drivewheel', 'Trim'])
new_data
```

<h2>Делаем все строки в lowercase, чтобы соотнести все значения</h2>

```python
new_data['Make'] = new_data['Make'].str.lower()
new_data['Model'] = new_data['Model'].str.lower()
new_data['Type'] = new_data['Type'].str.lower()
new_data
```

<h2>Запоняем пропуски в оставшихся столбцах с помощью группировок по стобцу Type, так как группировки нескольких столбцов не работают</h2>

```python
new_data.loc[:, 'Liter'] = new_data.groupby(['Type'])['Liter'].transform(lambda x: x.fillna(x.median()))
new_data
```

```python
new_data.loc[:, 'Mileage'] = new_data.groupby(['Type'])['Mileage'].transform(lambda x: x.fillna(x.median()))
new_data
```

```python
new_data.loc[:, 'Cruise'] = new_data.groupby(['Type'])['Cruise'].transform(lambda x: x.fillna(x.median()))
new_data
```

```python
new_data.loc[:, 'Sound'] = new_data.groupby(['Type'])['Sound'].transform(lambda x: x.fillna(x.median()))
new_data
```

```python
new_data.loc[:, 'Leather'] = new_data.groupby(['Type'])['Leather'].transform(lambda x: x.fillna(x.median()))
new_data
```

<h2>Посмотрим, что получилось с данными. Некоторые строки оказались пустыми</h2>

```python
new_data.describe()
```

<h2>Все строки с пустым значением Mileage имеют пустые значения в столбцах Liter, Cruise, Sound, Leather. Удалим эти строки</h2>

```python

new_data[new_data['Mileage'].isna()]

new_data.dropna(subset=['Mileage','Liter','Cruise','Sound', 'Leather'], how='all', inplace = True)
new_data.describe()
```

<h2>Проверим на дубликаты получившуюся таблицу</h2>

```python
new_data.duplicated().sum()
```

<h2>Посмотрим, какие типы в столбцах</h2>

```python
new_data.info()
```

<h2>Столбцы Cylinder и Doors оказались строчными, так как ранее в них были и строки, и числа. Преобразуем типы данных столбцов в float</h2>

```python
new_data['Cylinder'] = new_data['Cylinder'].astype(float)
new_data['Doors'] = new_data['Doors'].astype(float)
```

```python
new_data.info()
```

<h2>Посмотрим, какие значения есть в столбце  Make. Тут есть очень странные значения vw, maxda, vokswagen, toyouta. Скорее всего, тут была опечатка. Заменим эти значения на правильные - mazda, volkswagen, toyota. Остальные марки машин существуют</h2>

```python
new_data['Make'].value_counts()
```

```python
new_data.loc[new_data['Make'] == 'vw', 'Make'] = 'volkswagen'
new_data.loc[new_data['Make'] == 'vokswagen', 'Make'] = 'volkswagen' 
new_data.loc[new_data['Make'] == 'maxda', 'Make'] = 'mazda' 
new_data.loc[new_data['Make'] == 'toyouta', 'Make'] = 'toyota'
```

<h2>Разделим колонки на категориальные и числовые, произведем one-hot-encoding. Выделим отдельно столбец Price, так как он будет нужен для предсказаний</h2>

```python
num_cols = ['Mileage', 'Cylinder',	'Liter',	'Doors',	'Cruise',	'Sound',	'Leather']
cat_cols = ['Make', 'Model', 'Type']
dummy_features = pd.get_dummies(new_data[cat_cols])

Y = new_data['Price']
X = pd.concat([new_data[num_cols], dummy_features], axis=1)
X
```

<h2>Создаем модель линейной регрессии, обучаем ее на полученных данных. В качестве отклика берем столбец Price</h2>

```python
import numpy as np
import pandas as pd
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error



reg = LinearRegression().fit(X, Y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(X)
print('Error: {}'.format(mean_absolute_percentage_error(pred_values, Y)))
print('Error: {}'.format(mean_absolute_error(pred_values, Y)))
```

<h2>Попробуем удалить столбцы, чтобы улучшить предсказания</h2>

```python
new_data = new_data.drop(columns=['Leather', 'Sound', 'Cruise', 'Doors'])

num_cols = ['Mileage', 'Cylinder',	'Liter']
cat_cols = ['Make', 'Model', 'Type']
dummy_features = pd.get_dummies(new_data[cat_cols])

Y = new_data['Price']
X = pd.concat([new_data[num_cols], dummy_features], axis=1)

reg = LinearRegression().fit(X, Y)
print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(X)
print('Error: {}'.format(mean_absolute_percentage_error(pred_values, Y)))
print('Error: {}'.format(mean_absolute_error(pred_values, Y)))
```
