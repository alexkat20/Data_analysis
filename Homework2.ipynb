<h2>5. Убираем все nan'ы из колонки Income на среднее значение, потому что Income - это приход, и он может быть не целым. Поставим в пустые ячейки среднее значение.</h2>

```python
#data[data['Income'].isna()]

data['Income'] = data['Income'].fillna(data['Income'].mean())

data['Income'].isna().sum()
```

<h2>6. Убираем все nan'ы из колонки MntWines. Здесь поменяем все пропуски на медиану, так как среднее значение не подходит из-за того, что он не целое, а нам нужно целое значение в этой колонке</h2>

```python
#  data[data['MntWines'].isna()]
data['MntWines'] = data['MntWines'].fillna(data['MntWines'].median())
data['MntWines'].isna().sum()
```

<h2>7. Убираем все nan'ы из колонки MntFruits. Аналогично предудщему алгоритму, меняем все пустые ячейки на медиану, так как количество может быть только целым</h2>

```python
#  data[data['MntFruits'].isna()]
data['MntFruits'] = data['MntFruits'].fillna(data['MntFruits'].median())
data['MntFruits'].isna().sum()
```

<h2>8. Проверяем, остались ли пустые ячейки</h2>

```python
data.isna().sum()
```

<h2>13. Приведем колонки MntFruits и MntWines к целому количеству, так как количесвто не может быть дробным</h2>

<h2>14. Разделим все колонки на категориальные и числовые, чтобы далее удобнее было с ними работать и просмотрим, какие в них есть значения, чтобы найти странные или неправильные </h2>

```python

num_cols = ['Year_Birth', 'Income', 'Kidhome' , 'Teenhome', 'Recency',	'MntWines',	'MntFruits',	'MntMeatProducts',	'MntFishProducts',	'MntSweetProducts',	'MntGoldProds',	'NumDealsPurchases',	'NumWebPurchases',	'NumCatalogPurchases',	'NumStorePurchases',	'NumWebVisitsMonth',	'AcceptedCmp3',	'AcceptedCmp4',	'AcceptedCmp5',	'AcceptedCmp1',	'AcceptedCmp2',	'Complain',	'Z_CostContact',	'Z_Revenue',	'Response']


cat_col = ['Education', 'Marital_Status', 'Dt_Customer']

for col in cat_col:
    print(data[col].value_counts())

```

<h2>15. Оказалось, что в колонке Marital_Status есть странные значения  YOLO и Absurd. Они явно не подходят этим данным, поэтому их нужно убрать. Удаляем все колонки с такими значениями. Их оказалось совсем немного</h2>

```python
broken_indices = data.loc[data['Marital_Status'].isin(['Absurd', 'YOLO'])].index
data = data.drop(broken_indices)

data['Marital_Status'].value_counts()
```
